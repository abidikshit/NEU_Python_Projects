# Part 1.4

We can use a for loop to calculate the total inventory cost for different order quantities and then find the order quantity that results in the smallest total cost.

This code defines the demand, unit cost, holding cost rate, and order cost as variables. It then defines a range of order quantities to test using the range function. The for loop iterates over each order quantity in the range and calculates the total inventory cost using the mathematical functions defined in Point 2. The total cost for each order quantity is added to a list. The code then finds the order quantity that results in the smallest total cost using the min function and prints the optimal order quantity and total inventory cost.

```python
# Define the range of order quantities to test
order_quantities = range(100, 5000, 100)

# Calculate the total inventory cost for each order quantity
total_costs = []
for order_quantity in order_quantities:
    # Calculate the average inventory held throughout the year
    average_inventory = order_quantity / 2

    # Calculate the annual ordering cost
    annual_ordering_cost = annual_demand / order_quantity * cost_per_order

    # Calculate the annual holding cost
    annual_holding_cost = average_inventory * unit_cost * holding_cost_rate

    # Calculate the total inventory cost
    total_cost = annual_ordering_cost + annual_holding_cost

    # Add the total cost to the list
    total_costs.append(total_cost)

# Find the order quantity that results in the smallest total cost
min_cost = min(total_costs)
min_index = total_costs.index(min_cost)
optimal_order_quantity = order_quantities[min_index]

# Print the optimal order quantity and total inventory cost
print("Optimal order quantity:", optimal_order_quantity)
print("Total inventory cost:", min_cost)
```

```python
# Define the problem parameters
demand_mode = 15000
demand_min = 13000
demand_max = 17000
holding_cost = 0.144 * 80
ordering_cost = 220

# SD of minimum cost and Sample mean by constructing a 95% confidence interval

# Define a function to compute the optimal order quantity
def compute_optimal_order_quantity(annual_demand, holding_cost, ordering_cost):
    EOQ = np.sqrt((2 * annual_demand * ordering_cost) / holding_cost)
    return EOQ

# Define a function to compute the total cost
def compute_total_cost(annual_demand, order_quantity, holding_cost, ordering_cost):
    setup_cost = (annual_demand / order_quantity) * ordering_cost
    holding_cost = (order_quantity / 2) * holding_cost
    total_cost = setup_cost + holding_cost
    return total_cost

# Perform the simulation
min_costs = np.zeros(1000)
for i in range(1000):
    demand = np.random.triangular(demand_min, demand_mode, demand_max)
    order_quantity = compute_optimal_order_quantity(demand, holding_cost, ordering_cost)
    min_cost = compute_total_cost(demand, order_quantity, holding_cost, ordering_cost)
    min_costs[i] = min_cost

# Calculate the expected minimum total cost and the 95% confidence interval
mean_min_cost = np.mean(min_costs)
std_min_cost = np.std(min_costs, ddof=1)
n = len(min_costs)
z_critical = 1.96  # for 95% confidence interval
margin_of_error = z_critical * std_min_cost / np.sqrt(n)
lower_bound = mean_min_cost - margin_of_error
upper_bound = mean_min_cost + margin_of_error

# Print the results
print(f"Expected minimum total cost: {mean_min_cost:.2f}")
print(f"95% confidence interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Assuming 'min_costs' is the array of minimum cost values obtained from the simulation
sample_std = np.std(min_costs, ddof=1)

print("Sample Standard Deviation: ", sample_std)
```

```python
from scipy.stats import poisson
from scipy import stats

# Compute the expected number of orders per month
num_orders_per_month = 15000 / 12

# Generate a random sample of order quantities per month
sample_size = 1000
orders_per_month = np.random.poisson(num_orders_per_month, size=sample_size)

# Calculate the total number of orders per year
total_orders = orders_per_month * 12

# Construct a histogram
plt.hist(total_orders, bins=20)
plt.title('Histogram of Total Orders per Year')
plt.xlabel('Total Orders')
plt.ylabel('Frequency')
plt.show()

# Calculate the mean and standard deviation of the total number of orders
mean = np.mean(total_orders)
std = np.std(total_orders, ddof=1)

# Calculate the 95% confidence interval
conf_int = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(len(total_orders)))

print(f"Expected annual number of orders: {mean:.2f}")
print(f"95% Confidence interval: ({conf_int[0]:.2f}, {conf_int[1]:.2f})")
```
